```python
########################################################################################################################
# Load problem and set problem variables
########################################################################################################################
#problem_tsp='ulysses22'
#problem_tsp='aatemp'
#problem_tsp='h20'
problem_tsp='temp10'
if path.exists(str("./problems/"+problem_tsp+".tsp")):
	file_path = str("./problems/"+problem_tsp+".tsp")
else:
	sys.exit("ERROR: Problem file not found in ./problems/ :" + str (problem_tsp))
(tsp_name, tsp_df, tsp_dist_matrix) = tsp_load(file_path)	#load problem
no_of_cities=len(tsp_df.index.tolist())
print ("TSP laoded is         : " + tsp_name)
print ("No of cities is       : " + str (no_of_cities))
city_bits = math.ceil (math.log(no_of_cities,2))
tour_bits = city_bits * no_of_cities
print ("City bits = Log2(N)   = " + str (city_bits))
print ("Tour bits = N*Log2(N) = " + str (tour_bits))
```

```python
########################################################################################################################
# Set QEA variables
########################################################################################################################
#repair_type=1      	#type=1 random . type=2 greedy . type=3 none
Q_agents=city_bits      #No of quantum agents; working at 5 -30
C_agents=city_bits      #No of classical agents;
iterations=1000   	    #generation or iteration; working at 100
teta=0.001*np.pi    	    #Q gate
qubit_type=0		    #type=0 "qasm_simulator"; type=1 "ibmq_qasm_simulator";	type=2 IBM_Q_machine 
exec_time=0
penalty_type=0       	#type=0->no penalty, type=1->Add one max edge len for each error
dev_name='least_busy'
verbose=1               #0->min, 1->plot live value; 2->high

print ("Total agents = Q_agents * C_agents = " + str (Q_agents) + " * " + str (C_agents) + " = " + str (Q_agents * C_agents))
```

```python
########################################################################################################################
# Run QEA algo
########################################################################################################################
iteration_fitness, best_tours, exec_time, Q_pop = TSP_quantum (Q_agents, C_agents, tsp_dist_matrix, city_bits, no_of_cities, iterations, teta, qubit_type, exec_time, penalty_type, dev_name, verbose)
```

```python
########################################################################################################################
# Plot graph
########################################################################################################################
## Plot fitness values of Q_agents
iteration_fitness_df = pd.DataFrame(np.matrix(iteration_fitness))   
iteration_fitness_df.plot()  
plt.show() 
```
